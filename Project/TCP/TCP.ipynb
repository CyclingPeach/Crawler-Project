# html

```python
import requests

tcp_url = 'http://www.rfc-editor.org/rfc/rfc793.txt'
html = requests.get(tcp_url).text
print(html)
```

# parsel

```python
import requests
from parsel import Selector

tcp_url = 'http://www.rfc-editor.org/rfc/rfc793.txt'

res = requests.get(tcp_url)
print(res.status_code)
# rint(res.text)

selector = Selector(res.text)
content = selector.css('p').getall()[0]
print(content)
```

# BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

tcp_url = 'http://www.rfc-editor.org/rfc/rfc793.txt'
html = requests.get(tcp_url).text
soup = BeautifulSoup(html)
print(soup.p)
```

# Lxml

```python
import requests
from lxml import etree

tcp_url = 'http://www.rfc-editor.org/rfc/rfc793.txt'
res = requests.get(tcp_url)
html = etree.HTML(res.text)
print(html.xpath('//p/text()')[0])
```
